```python
# GOAL: maximize likelihood of the data w.r.t. model parameters (the probabilities - statistical modeling)
#   So, it is the product of all the probabilities. This will be very small 
#   numbers less than 1. So, for convenience, we transform as follows:
# equivalent to maximizing the log likelihood (because log is monotonic)
# equivalent to minimizing the negative log likelihood
# equivalent to minimizing the average negative log likelihood

# log(a*b*c) = log(a) + log(b) + log(c)
```

```python
log_likelihood = 0.0
n=0
# for w in ['andrejq']:
for w in words[:3]:
    chs = ['.'] + list(w) + ['.']
    # print(chs)
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        prob = P[ix1, ix2]
        logprob = torch.log(prob)
        log_likelihood += logprob
        n+=1
        print(f'{ch1}{ch2}: {prob:.4f} {logprob:.4f}')
        
print(f'{log_likelihood=}')
# nll - negative log likelihood
nll = -log_likelihood
print(f'{nll=}')
# the lower the number, the better
print(f'{nll/n}')
```

```python
# We need 27 neurons as the one-hot gave 27 inputs. 
# Below is with 27 neurons - it evaluates the 27 neurons with 
# the 5 character inputs.
W = torch.randn(27, 27)
xenc @ W # 5 x 27 matrix
# each element below tells us what the firing rate is for every one 
# of the five examples. For example, the element at [3,13] tells us
# that the firing rate of the 13th neuron looking at the 3rd input.
# It was achieved by a dot-product of the 3rd input and the 13th 
# column of the W matrix.
# In summary, using matrix multiplication, we can very efficiently evaluate
# the dot-product between lots of input examples and lots of neurons
# (each column in W which has the weights of each neuron)
```

```python
(xenc @ W)[3,13]
```

```python
# we fed 27 inputs (xenc) to 27 neurons in the first layer of a neural network
# The neurons perform w * x and bias b is zero (no bias). The neurons also
# don't have a non-linearity operation at the end like tanh. So, it is going
# to be just a linear layer. In addition to that, we won't have other layers.
# Simplest Neural Net - a single linear layer.
#
# What do we want the outputs to be? - we would like to produce the probability
# distribution for the next character - like the one above that we did with counts. 
# 
# to do it, we interpret the numbers above by using log counts and exponentiating
# them - the negative numbers become positive numbers below 1 and positive
# numbers become numbers above 1.
# 
# These nummbers can now take different values based on the settings 
# we give for W.
#
# 
# further, normalizing them gives us probabilities - i.e., what is the 
# probability that each one of the 27 characters are likely to come next.
#
# As we tune the weights, we get different probabilties out.
logits =xenc @ W  # logits -> log counts
counts = logits.exp() # equivalent to N above.
probs = counts / counts.sum(1, keepdim=True)
probs
# All of these operations are differentiable and so we can 
# back-propogate and do updates to nudge the weights in the 
# direction of the gradient to optimize.
# So, can we optimize W so that the probabilities coming out
# are good which is measured by the loss function.
```

```python
# forward pass
xenc = F.one_hot(xs, num_classes=27).float() # input to the network: one-hot encoding
logits = xenc @ W # predict log-counts
counts = logits.exp() # counts, equivalent to N
probs = counts / counts.sum(1, keepdims=True) # probabilities for next character
# the last two lines are called a softmax - it outputs a probability distribution 
# (sums to 1) out of outputs from a neuron which can be positive or negative.
# softmax makes the neural net outputs probabilities. 
# the operations done by the neural network layer are all differentiable
# and so we can back-propagate.
```

```python
# example showing how good the probabilities 
# produced by the neural net are
nlls = torch.zeros(5)
for i in range(5):
    # i-th bigram:
    x = xs[i].item() # input character index
    y = ys[i].item() # label character index
    print('-----------')
    print(f'bigram example {i+1}: {itos[x]}{itos[y]} (indexes {x}, {y})')
    print('input to the neural net:', x)
    print('output probabilities from the neural net:', probs[i])
    print('label (actual next character):', y)
    p = probs[i, y]
    print('probability assigned by the net to the correct characters:', p.item())
    logp = torch.log(p)
    print('log likelihood', logp.item())
    nll = -logp
    print('negative log likelihood:', nll.item())
    nlls[i] = nll
    
print('***************')
print('average negative log likelihood, i.e. loss = ', nlls.mean().item())


    
```

```python
# forward pass
xenc = F.one_hot(xs, num_classes=27).float() # input to the network: one-hot encoding
logits = xenc @ W # predict log-counts
counts = logits.exp() # counts, equivalent to N
probs = counts / counts.sum(1, keepdims=True) # probabilities for next character
loss = -probs[torch.arange(5), ys].log().mean()
```

```python
# backward pass
W.grad = None # gradient set to zero
loss.backward()
# fills in the gradients all the way back to the W (parameters of the neural net)
# W.grad is telling us the influence of those weights on the loss function
```

```python
# update the tensor by nudging the weights in the opposite direction of the gradient by a small amount
W.data += -0.1 * W.grad
```

```python
# create the dataset
xs, ys = [], []

for w in words:
    chs = ['.'] + list(w) + ['.']
    # print(chs)
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        # print(ch1, ch2)
        xs.append(ix1)
        ys.append(ix2)
xs = torch.tensor(xs)
ys = torch.tensor(ys)
num = xs.nelement()
print('number of examples: ', num)

# initialize the 'network'.
g = torch.Generator().manual_seed(2147483647)
W = torch.randn((27, 27), generator=g, requires_grad=True)
```

```python
# gradient descent
for k in range(100):
    
    # forward pass
    xenc = F.one_hot(xs, num_classes=27).float() # input to the network: one-hot encoding
    logits = xenc @ W # predict log-counts
    counts = logits.exp() # counts, equivalent to N
    probs = counts / counts.sum(1, keepdims=True) # probabilities for next character
    # Note: the 2nd term is regularization term. See Explanation in Note 2 below. 
    loss = -probs[torch.arange(num), ys].log().mean() + 0.01*(W**2).mean()
    print(loss.item())
    
    # backward pass
    W.grad = None # gradient set to zero
    loss.backward()
    
    # update
    W.data += -50 * W.grad
    
```

```python
# 1:45:00 the neural net approach is scalable but the counting approach
# is not. The bigrams used only needed to keep track of one character and 
# so we could use a table for the counts. To predict the next character from
# a sequence of characters, the table size would grow exponentially 27 x 27 x ... rows.
# For neural nets, the forward pass will get more complicated with more layers
# etc. but it is scalable.
```

```python
# NOTE 1: Minute 1:47:45: xenc is made up of ONE-HOT VECTORS which are 
#       multiplied by W matrix. We think of this as multiple neurons 
#       being forwarded in a fully connected manner. But, what is happening
#       is the following: say, you have one-hot vector with 1 at the 5th 
#       dimension. The matrix multiplication with W picks up the 5th row of W.
#       i.e., logits will become the 5th row of W. This is similar to the way
#       we looked up the next character by indexing using the first character.
#       So, the W (exponentiated) is literally the same as the 
#       2-dimensional array used for counting.
#       As we apply gradient descent to optimize the loss function, we end
#       up with similar results as well.
```

```python
# NOTE 2: Minute 1:50: if we increase the count in the smoothing operation done
#         with (N+count).float(), probability becomes more uniform. 
#         Gradient-based framework has an equivalent to smoothing - 
#         make W all equal to each other. If you incentivize W through
#         the loss function, the more uniform the distribution will be.
#         REGULARIZATION: we augment the loss function to have a small 
#         component called the regularization loss. For example, we could 
#         take W, square the entries, and then mean to get positive numbers. 
#         Adding a regularization term to the loss (line 9 above) results
#         in two components for the loss: it tries to make all
#         the probabilities work out. Also, it simultaneously tries to 
#         make all W's be zero as there will be an additional loss if they
#         are non-zero. So, increasing the count in the smoothing operation
#         corresponds to increasing the regularization term 0.01 (line 9).
```

```python
# finally, sample from the 'neural net' model
g = torch.Generator().manual_seed(2147483647)

for i in range(5):
    out = []
    ix = 0
    while True:
        
        # --------------
        # BEFORE:
        # p = P[ix]
        # NOW:
        xenc = F.one_hot(torch.tensor([ix]), num_classes=27).float()
        logits = xenc @ W # predict log-counts
        counts = logits.exp() # counts, equivalent to N
        p = counts / counts.sum(1, keepdims=True) # probabilities for next character
        # --------------
        
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```
